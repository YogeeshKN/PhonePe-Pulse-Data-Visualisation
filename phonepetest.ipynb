```python
import os
import json
import pandas as pd
import mysql.connector as sql
```

```python
# aggregated folder details to be fetched and converted to data frame
# aggregator_transaction

path1 = "E:/GUVI Projects/Phonepe Pulse/pulse/data/aggregated/transaction/country/india/state/"
aggr_tran_list = os.listdir(path1)


columns1 = {"States":[], "Years":[], "Quarter":[], "Transaction_type":[], "Transaction_count":[], "Transaction_amount":[]}
for state in aggr_tran_list:
    current_states = path1+state+"/" # the end / is to select the folders within the state folder
    aggr_year_list = os.listdir(current_states)
    
    for year in aggr_year_list:
        current_year = current_states +year+"/"
        aggr_file_list = os.listdir(current_year)   # creating the loop to load the csv files.
                                                  # one by one loading is difficult so creating a loop for uploading
        for file in aggr_file_list:
            current_file = current_year + file
            data = open(current_file, "r")

            A = json.load(data)
            for i in A["data"]["transactionData"]:
                name = i['name']
                count = i["paymentInstruments"][0]["count"]
                amt = i["paymentInstruments"][0]["amount"]
                columns1["Transaction_type"].append(name)
                columns1["Transaction_count"].append(count)
                columns1["Transaction_amount"].append(amt)
                columns1["States"].append(state)
                columns1["Years"].append(year)
                columns1["Quarter"].append(int(file.strip(".json")))
```

```python
aggre_transaction = pd.DataFrame(columns1) # converting it to df so that neatly represented in tables.

aggre_transaction
```

```python
# aggregator_user
path2 = "E:/GUVI Projects/Phonepe Pulse/pulse/data/aggregated/user/country/india/state/"
aggr_user_list = os.listdir(path2)


columns2 = {"States":[], "Years":[], "Quarter":[], "Brands":[], "Transaction_count":[], "Percentage":[]}
for state in aggr_user_list:
    current_states = path2+state+"/" # the end / is to select the folders within the state folder
    aggr_year_list = os.listdir(current_states)
    
    for year in aggr_year_list:
        current_year = current_states +year+"/"
        aggr_file_list = os.listdir(current_year)   # creating the loop to load the csv files.
                                                  # one by one loading is difficult so creating a loop for uploading
        for file in aggr_file_list:
            current_file = current_year + file
            data = open(current_file, "r")

            B = json.load(data)
            
            try:
                for i in B["data"]["usersByDevice"]:
                    brand = i["brand"]
                    count = i["count"]
                    percentage = i["percentage"]
                    columns2["Brands"].append(brand)
                    columns2["Transaction_count"].append(count)
                    columns2["Percentage"].append(percentage)
                    columns2["States"].append(state)
                    columns2["Years"].append(year)
                    columns2["Quarter"].append(int(file.strip(".json")))
            
            except:
                pass
```
